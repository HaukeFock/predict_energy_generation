# Creating functions for dataframe generation

```python
def clean_dataframe(df):
    df.drop(columns=["Produkt_Code", "Qualitaet_Byte","Qualitaet_Niveau"],inplace=True)
    df.rename(columns={'Zeitstempel':'Date','SDO_ID':'Station_ID','Wert':'Value'},inplace=True)
    df.Station_ID = df.Station_ID.astype('int32')
    if df.Value.dtypes == 'float64':
        df.Value = df.Value.astype('float32')
    if df.Value.dtypes == 'int64':
        df.Value = df.Value.astype('int32')
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.pivot(index="Date", columns="Station_ID", values="Value")
    return df
```
