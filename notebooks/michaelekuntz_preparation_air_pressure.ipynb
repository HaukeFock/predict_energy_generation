```python
hourly_airp_df = pd.read_csv("/Users/michael.kuntz/code/HaukeFock/predict_energy_generation/raw_data/Features/Air pressure/Hourly/data/data_OBS_DEU_PT1H_T2M.csv", delimiter = ",",index_col=False)
hourly_airp_df.drop(columns=["Produkt_Code", "Qualitaet_Byte","Qualitaet_Niveau"],inplace=True)

hourly_airp_df.head()
```

```python
hourly_airp_df.rename(columns={'Zeitstempel':'Date','SDO_ID':'Station_ID','Wert':'Value'},inplace=True)
```

```python
hourly_airp_df['Date'] = pd.to_datetime(hourly_airp_df['Date'])
hourly_airp_df.head()
```

```python
daily_airp_df = pd.read_csv("/Users/michael.kuntz/code/HaukeFock/predict_energy_generation/raw_data/Features/Air pressure/Daily/data/data_OBS_DEU_P1D_P0.csv", delimiter = ",",index_col=False)
daily_airp_df.drop(columns=["Produkt_Code", "Qualitaet_Byte", "Qualitaet_Niveau"],inplace = True)
daily_airp_df.rename(columns={'Zeitstempel':'Date','SDO_ID':'Station_ID','Wert':'Value'},inplace=True)
daily_airp_df['Date'] = pd.to_datetime(daily_airp_df['Date'])
daily_airp_df.head()
```

# Creating pivot tables with all station_IDs as columns

```python
hourly_airp_df = hourly_airp_df.pivot(index="Date", columns="Station_ID", values="Value")
hourly_airp_df.head()
```

```python
daily_airp_df = daily_airp_df.pivot(index="Date", columns="Station_ID", values="Value")
daily_airp_df.head()
```
